```python
try:
    url = 'https://www.marketwatch.com/investing/stock/goev/financials?mod=mw_quote_tab'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find("table", class_='table table--overflow align--right')
    table_row = soup.find("tbody", class_='table__body row-hover').find_all('tr')

    with open("canoo_income_statement.csv", "w", newline="") as csv_file:
        csv_writer = csv.writer(csv_file)
        
        # Write header
        csv_writer.writerow(["item", "2018", "2019", "2020", "2021",'2022','nan'])

        for data in table_row:
            rows = data.find_all('td')
            row = []
            for j in rows:
                val = j.find_all('div', class_='cell__content')[0]
                row.append(val.get_text(strip=True))
            csv_writer.writerow(row)
            print(row)

except Exception as e:
    print(f"Error: {e}")

print("CSV file created successfully.")
```

```python
canoo_income_statement = pd.read_csv('canoo_income_statement.csv').drop('nan',axis=1)
rivian_income_statement = pd.read_csv("rivian_income_statement.csv").drop("None", axis=1)
tesla_income_statement = pd.read_csv("tesla_income_statement.csv").drop("None", axis=1)
```

```python
def preprocess_(x):
    try:
        if x == '-':
            return 0
        else:
            return x
    except Exception as e:
        print('Error in preprocess_:', e)
```

```python
def remove_paranthesis(x):
    try:
        return str(x).replace("(", "").replace(")", "")
    except Exception as e:
        print('Error in remove_paranthesis:', e)
```

```python
def percentage(x):
    try:
        if isinstance(x, str):
            val = x[-1]
            if val == "%":
                x = x.replace(',', '')[:-1]
                return float(x) / 100
        return x
    except Exception as e:
        print('Error in percentage:', e)  
```

```python
def replace_digits(x):
    try:
        x = str(x)
        val = x[-1].lower()
        if val.isdigit(): 
            return float(x)
        elif val == 'k':
            return float(x[:-1]) * 1000
        elif val == 'm':
            return float(x[:-1]) * 1000000
        elif val == 'b':
            return float(x[:-1]) * 1000000000
        else:
            return float(x)
    except Exception as e:
        print('Error in replace_digits:', e)
```

```python
def preprocess_df(column):
    try:
        column = column.apply(preprocess_)
        column = column.apply(remove_paranthesis)
        column = column.apply(percentage)
        column = column.apply(replace_digits)
        return column
    except Exception as e:
        print('Error in preprocess_df:', e)
```

```python
for id,column in enumerate(canoo_income_statement):
    if id >=1:
        canoo_income_statement[column] = preprocess_df(canoo_income_statement[column])
    else:
        pass
```
